```python
# 对处理后的文件进行分词
import jieba
def cut_words(sentence):
    return " ".join(jieba.cut(sentence)).encode('utf-8')
```

```python
import codecs
f=codecs.open(r'./wiki_zh_simple.txt','r',encoding="utf8")
target = codecs.open(r"./wiki_zh_simple_cut.txt", 'w',encoding="utf8")
```

```python
lines = f.readlines()
corpus = []
for i ,line in enumerate(lines):
    if i %1000==0:
         print('---- processing ', i, ' article----------------')
    words = list(jieba.cut(line.strip()))
    if words != '':
        corpus.append(words)
    
```

```python
corpus[:2]
```

```python
lines[:2]
```

```python
# 构建词向量
from gensim.models import word2vec
model = word2vec.Word2Vec(corpus, size=100, window=20, min_count=200, workers=8)
```

```python
model.wv['数学']
```

```python
model.wv.most_similar('苹果')
```

1、遇到训练模型运行速度特别慢
   UserWarning: C extension not loaded, training will be slow. Install a C compiler and reinstall gensim for fast training.
  "C extension not loaded, training will be slow. "
解决方法：后面查看相关文档说是，pip install gensim安装方式会缺少C扩展库，卸载，用官网的conda install -c conda-forge gensim的方式重新安装了，速度快很多。应该还是底层的问题吧
